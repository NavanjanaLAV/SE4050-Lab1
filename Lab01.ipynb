```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate from exponential distribution
arr_4x4 = np.random.exponential(scale=1.0, size=(4,4))
print(arr_4x4)
```

```python
# 100,000 values
exp_data = np.random.exponential(scale=1.0, size=(100000,))

# Plot alongside uniform and normal distributions
plt.hist(exp_data, density=True, bins=100, histtype="step", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", label="Uniform")
plt.hist(np.random.normal(size=100000), density=True, bins=100, histtype="step", label="Normal")

plt.legend()
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
# Assuming df is from seaborn tutorial dataset
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

pearson_corr = df[features].corr(method='pearson')
spearman_corr = df[features].corr(method='spearman')

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation')

plt.subplot(1,2,2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title('Spearman Correlation')
plt.show()
```

```python
df = pd.read_csv("Au_nanoparticle_dataset .csv")
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
print(new_df.describe())  # includes mean, std, 25%, 50%, 75%
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16,4))
for ax, col in zip(axes, new_df.columns):
    ax.hist(new_df[col], bins=20, color='skyblue')
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot, fill=True)
plt.show()
```
